# Hello DSPy: raw prompts vs signatures

This example shows the contrast at the heart of Chapter 1 — writing prompts as strings versus declaring a signature and letting DSPy build the prompt for you.

## Setup

Set your `OPENAI_API_KEY` in a `.env` file at the repo root before running the cell below.

```
import os
import dspy
from dotenv import load_dotenv

load_dotenv()

dspy.configure(lm=dspy.LM('openai/gpt-5-mini'))
```

## Without DSPy: a raw prompt string

Traditional prompting means brittle string replacement: you write the template, then `.format(...)` the variables in by hand.

```
prompt = """You’re an expert in {topic}. Given context: {context}, answer: {question}"""
formatted_prompt = prompt.format(
    topic="Physics",
    context="The speed of light is 300,000 km/s.",
    question="What is the speed of light?",
)

print(formatted_prompt)
```

## With DSPy: a signature

In DSPy you declare the inputs and outputs as a class, and DSPy handles building the prompt for you.

```
class QA(dspy.Signature):
    """Answer the question as an expert in the topic."""
    topic = dspy.InputField()
    context = dspy.InputField()
    question = dspy.InputField()
    answer = dspy.OutputField()

qa = dspy.Predict(QA)

result = qa(
    topic="Physics",
    context="The speed of light is 300,000 km/s.",
    question="What is the speed of light?",
)

print(result.answer)
```

## See the prompt DSPy generates

Use `dspy.inspect_history` to view the fully assembled prompt and response from the most recent call.

```
dspy.inspect_history(n=1)
```
