````python
import os
import dspy
from dotenv import load_dotenv

load_dotenv()

dspy.configure(lm=dspy.LM('openai/gpt-5-mini'))
````

## A portfolio calculator

Instead of asking the model to reason through a calculation, ask it to write the code that does it.


````python
import dspy

calculator = dspy.ProgramOfThought(
    "portfolio: str -> total_return: str"
)

result = calculator(
    portfolio="AAPL: bought 50 shares at $142, now $198. "
              "GOOGL: bought 20 shares at $2,840, now $2,950. "
              "MSFT: bought 30 shares at $310, now $378."
)
print(result.total_return)
````
